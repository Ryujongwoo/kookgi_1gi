```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
import time
```

```python
# 노래 제목 크롤링
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20220712&hh=12&rtm=Y&pg='
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
for i in range(2, 3):
    targetSite = '{}{}'.format(url, i)
    req = requests.get(targetSite, headers=header)
    # print(req)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    # print(soup)
    
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    # print(titles)
    # print(len(titles))
    
    for i in range(len(titles)):
        # print('{:3d}위 {}'.format(i + 1, titles[i].text.strip()))
        # print('{:3d}위 {}'.format(i + 1, titles[i].text.split('\n')[-1].strip()))
        print('{:3d}위 {}'.format(i + 1, titles[i].text.split('\n')[-1].strip()))
```

```python
# 아티스트 크롤링
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20220712&hh=12&rtm=Y&pg='
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
for i in range(1, 2):
    targetSite = '{}{}'.format(url, i)
    req = requests.get(targetSite, headers=header)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    # print(artists)
    # print(len(artists))
    
    # i = 1
    # for artist in artists[5:]:
        # print('{:3d}위 {}'.format(i, artist.text.strip()))
        # i += 1
        
    # for i in range(len(artists[5:])):
        # print('{:3d}위 {}'.format(i + 1, artists[i + 5].text.strip()))
        
    for i in range(5, len(artists)):
        # print('{:3d}위 {}'.format(i - 4, artists[i].text.strip()))
        # print('{:3d}위 {}'.format(i - 4, artists[i].text.split('\n')[0]))
        print('{:3d}위 {}'.format(i - 4, artists[i].text.strip().split('\n')[0]))
```

```python
# 크롤링한 4페이지의 결과를 리스트 하나에 모은다.
genie = []
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20220712&hh=12&rtm=Y&pg='
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    req = requests.get(targetSite, headers=header)
    time.sleep(0.5)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    
    for j in range(50):
        genie.append(
            (artists[j + 5].text.strip().split('\n')[0], titles[j].text.split('\n')[-1].strip())
        )
# print(len(genie))
print(genie)
```

```python
print(dt.now())
for i in range(200):
    print('{:3d}위 {} - {}'.format(i + 1, genie[i][0], genie[i][1]))
```

```python
with open('./output/genieTOP200.txt', 'w', -1, 'utf-8') as file:
    file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
    for i in range(200):
        file.write('{:3d}위 {} - {}\n'.format(i + 1, genie[i][0], genie[i][1]))
print('genieTOP200.txt 파일로 쓰기 완료')
```
