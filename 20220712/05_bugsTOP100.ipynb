```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
print(dt.now())
```

```python
targetSite = 'https://music.bugs.co.kr/chart'
req = requests.get(targetSite)
# print(req)
html = req.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll('p', {'class': 'title'})
# print(titles)
# print(len(titles))

for i in range(len(titles)):
    # print('{:3d}위 {}'.format(i + 1, titles[i].text.strip()))
    # print('{:3d}위 {}'.format(i + 1, titles[i].text.split('\n')[1]))
    print('{:3d}위 {}'.format(i + 1, titles[i].text.strip().split('\n')[0]))
```

```python
# 아티스트 크롤링
artists = soup.findAll('p', {'class': 'artist'})
# print(artists)
# print(len(artists))

for i in range(len(artists)):
    # print('{:3d}위 {}'.format(i + 1, artists[i].text.strip())) # 적절 X
    # print('{:3d}위 {}'.format(i + 1, artists[i].text.split('\n')[1])) # 적절 X
    print('{:3d}위 {}'.format(i + 1, artists[i].text.strip().split('\n')[0]))
```

```python
print(dt.now().strftime('%Y-%m-%d %H:%M:%S'))
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{:3d}위 {} - {}'.format(i + 1, artist, title))
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다.
today = dt.now()
# file = open('C:/kookgi/venv/workspace_PythonCrawling/output/bugsTOP100.txt', 'w') # 절대경로
# file = open('./output/bugsTOP100.txt', 'w') # 상대경로
filename = './output/bugsTOP100_' + today.strftime('%Y%m%d%H%M%S') + '.txt'
file = open(filename, 'w', -1, 'utf-8')
file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    file.write('{:3d}위 {} - {}\n'.format(i + 1, artist, title))
file.close()
print(filename[9:] + ' 파일로 쓰기 완료')
```

```python
today = dt.now()
filename = './output/bugsTOP100_' + today.strftime('%Y%m%d%H%M%S') + '.txt'
# with 구문을 사용해 파일을 open 하면 with 블록의 내용이 모두 실행된 후 자동으로 파일을 닫는다.
with open(filename, 'w', -1, 'utf-8') as file:
    file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
    for i in range(100):
        artist = artists[i].text.strip().split('\n')[0]
        title = titles[i].text.strip().split('\n')[0]
        file.write('{:3d}위 {} - {}\n'.format(i + 1, artist, title))
print(filename[9:] + ' 파일로 쓰기 완료')
```
